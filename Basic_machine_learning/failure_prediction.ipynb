```python
import pandas as pd

df=pd.read_csv('test_data_truck.csv')
df
```

```python
df.shape
df.describe()
# good no empty/none column
```

```python
# Just to ensure no invalid data in faiure column
X= df.drop(columns =['date','failure'])
Y = df['failure']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import numpy as np
import joblib
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.2)
model =DecisionTreeClassifier()
model.fit(X_train, Y_train)
# lets see what would be the failure status of truck no. 34, 796 with some random oil properties and other parameters and no training of the data
prediction =model.predict([ [5.0 , 1.5, 10.4, 28.3, 110, 50 ,34, 885.01, 2470.8] , [9.0 , 41.5, 355.4, 1228.3, 1110, 90 ,796, 985.01, 1470.8] ])
#prediction
better_prediction = model.predict(X_test) 
better_prediction
score = accuracy_score(Y_test,better_prediction )
score
# accuracy looks good from .99 to 1.0
joblib.dump(model,'failure-predictions.joblib')
tree.export_graphviz(model, out_file ='failure_prediction.dot',
                     feature_names =['si','fe','al','particulate','temp','avg_speed','id','colour','viscosity'],
                     class_names= np.array(Y.unique()).astype('str').tolist(),
                     label = 'all',
                     rounded =True,
                     filled=True)
```
